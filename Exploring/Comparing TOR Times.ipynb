```python
import pandas as pd
```

```python
preprocessed_TOTs = pd.read_csv("../AdVitam/Exp2/Preprocessed/Physio and Driving/timestamps_obstacles.csv")
preprocessed_TOTs.iloc[:, 2:] = preprocessed_TOTs.iloc[:, 2:].apply(pd.to_timedelta, unit="s")
preprocessed_TOTs["subject_id"] = preprocessed_TOTs["subject_id"].apply(lambda x: x.split("T")[0] + "T" + x.split("T")[1].zfill(2))
preprocessed_TOTs["subject_id"] = preprocessed_TOTs["subject_id"].astype(str)
preprocessed_TOTs.drop(columns=["label_st"], inplace=True)
preprocessed_TOTs.sort_values(by=["subject_id"], inplace=True)
preprocessed_TOTs.set_index("subject_id", inplace=True)
preprocessed_TOTs
```

```python
manually_processed_TOTs = pd.read_csv("../Physiological Preprocessed/Exp2/driver_takeover_times.csv")
manually_processed_TOTs.rename(columns={"Unnamed: 0": "subject_id"}, inplace=True)
manually_processed_TOTs.iloc[:, 2:] = manually_processed_TOTs.iloc[:, 2:].apply(pd.to_timedelta)
manually_processed_TOTs.sort_values(by="subject_id", inplace=True)
manually_processed_TOTs.set_index("subject_id", inplace=True)
manually_processed_TOTs
```

```python
preprocessed_TOTs["TrigObsDeer"] = pd.to_timedelta(preprocessed_TOTs["TrigObsDeer"])
manually_processed_TOTs["TriggeredObs1"] = pd.to_timedelta(manually_processed_TOTs["TriggeredObs1"])

triggers = manually_processed_TOTs["TriggeredObs1"].apply(lambda x: x.total_seconds()) - preprocessed_TOTs["TrigObsDeer"].apply(lambda x: x.total_seconds())
triggers = triggers.to_frame()
triggers.columns = ["TrigObsDeer - TriggeredObs1"]
triggers["TrigObsDeer - TriggeredObs1"] = pd.to_timedelta(triggers["TrigObsDeer - TriggeredObs1"], unit="s")
triggers
```
